```python
grouped_df = df.groupby('trip_id')
```

```python
trip_start_time = grouped_df['arrival_time'].min()
```

```python
trip_end_time = grouped_df['arrival_time'].max()
```

```python
start_stop = grouped_df['stop_sequence'].min()
```

```python
trip_start_time_seconds = trip_start_time.str.split(':')
trip_end_time_seconds = trip_end_time.str.split(':')

for i,time in enumerate(trip_start_time_seconds):
    seconds = 60*60*int(time[0]) + 60*int(time[1]) + int(time[2])
    trip_start_time_seconds.iloc[i] = seconds

for i,time in enumerate(trip_end_time_seconds):
    seconds = 60*60*int(time[0]) + 60*int(time[1]) + int(time[2])
    trip_end_time_seconds.iloc[i] = seconds
    
trip_start_time_seconds = trip_start_time_seconds.astype('int64')
trip_end_time_seconds = trip_start_time_seconds.astype('int64')

trip_length = trip_end_time_seconds - trip_start_time_seconds
```

```python
df_trip = pd.DataFrame({'start_time':trip_start_time,'end_time':trip_end_time,'trip_length':trip_length,'stop_sequence':start_stop})
```

```python
df_trip.head()
```

```python
df_trip = df_trip.merge(df,on=['trip_id','stop_sequence'])
```
